```python
mean_ALQ101 = np.mean(data['ALQ101'])
mean_ALQ101

data['ALQ101'].replace(np.nan, mean_ALQ101, inplace=True)

data.isnull().sum()
```

```python
mean_ALQ110 = np.mean(data['ALQ110'])
mean_ALQ130 = np.mean(data['ALQ130'])
mean_DMDCITZN = np.mean(data['DMDCITZN'])
mean_DMDEDUC2 = np.mean(data['DMDEDUC2'])
mean_DMDMARTL = np.mean(data['DMDMARTL'])
mean_INDFMPIR = np.mean(data['INDFMPIR'])
mean_BPXSY1 = np.mean(data['BPXSY1'])

data['ALQ110'].replace(np.nan, mean_ALQ110, inplace=True)
data['ALQ130'].replace(np.nan, mean_ALQ130, inplace=True)
data['DMDCITZN'].replace(np.nan, mean_DMDCITZN, inplace=True)
data['DMDEDUC2'].replace(np.nan, mean_DMDEDUC2, inplace=True)
data['DMDMARTL'].replace(np.nan, mean_DMDMARTL, inplace=True)
data['INDFMPIR'].replace(np.nan, mean_INDFMPIR, inplace=True)
data['BPXSY1'].replace(np.nan, mean_BPXSY1, inplace=True)

data.isnull().sum()
```

```python
data.info()
```

```python
mean_BPXDI1 = np.mean(data['BPXDI1'])
mean_BPXSY2 = np.mean(data['BPXSY2'])
mean_BPXDI2 = np.mean(data['BPXDI2'])
mean_BMXWT = np.mean(data['BMXWT'])
mean_BMXHT = np.mean(data['BMXHT'])

data['BPXDI1'].replace(np.nan, mean_BPXDI1, inplace=True)
data['BPXSY2'].replace(np.nan, mean_BPXSY2, inplace=True)
data['BPXDI2'].replace(np.nan, mean_BPXDI2, inplace=True)
data['BMXWT'].replace(np.nan, mean_BMXWT, inplace=True)
data['BMXHT'].replace(np.nan, mean_BMXHT, inplace=True)

data.isnull().sum()
```

```python
mean_BMXBMI = np.mean(data['BMXBMI'])
mean_BMXLEG = np.mean(data['BMXLEG'])
mean_BMXARML = np.mean(data['BMXARML'])
mean_BMXARMC = np.mean(data['BMXARMC'])
mean_BMXWAIST = np.mean(data['BMXWAIST'])
mean_HIQ210 = np.mean(data['HIQ210'])

data['BMXBMI'].replace(np.nan, mean_BMXBMI, inplace=True)
data['BMXLEG'].replace(np.nan, mean_BMXLEG, inplace=True)
data['BMXARML'].replace(np.nan, mean_BMXARML, inplace=True)
data['BMXARMC'].replace(np.nan, mean_BMXARMC, inplace=True)
data['BMXWAIST'].replace(np.nan, mean_BMXWAIST, inplace=True)
data['HIQ210'].replace(np.nan, mean_HIQ210, inplace=True)

data.isnull().sum()
```
